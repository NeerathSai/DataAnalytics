```python
df=pd.DataFrame([('Neerath',1,'M',40,45,30),
                ('Sai',2,'M',30,35,40),
                ('Reddy',3,'M',25,50,30),
                ('Peggy',4,'F',45,40,50),
                ('Pepper',5,'F',40,40,40),
                ('Rose',6,'F',30,35,25),
                ('Chris',7,'M',20,25,20),
                ('Emily',8,'F',30,25,35),
                ('Ajak',9,'F',15,25,35),
                ('Tom',10,'M',40,40,45)],
               columns=('Name','RollNo','Gender','Marks1','Marks2','Marks3'))
df
# Data Set of 10 students with attributes like: Name, RollNo, Gender and Marks
```
