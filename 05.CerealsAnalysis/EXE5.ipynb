```python
df['protein'] =df['protein'].fillna(df['protein'].mean())
df['fat'] =df['fat'].fillna(df['fat'].mean())
df['sodium'] =df['sodium'].fillna(df['sodium'].mean())
df['fiber'] =df['fiber'].fillna(df['fiber'].mean())
df['carbo'] =df['carbo'].fillna(df['carbo'].mean())
df['sugars'] =df['sugars'].fillna(df['sugars'].mean())
df['potass'] =df['potass'].fillna(df['potass'].mean())
df['vitamins'] =df['vitamins'].fillna(df['vitamins'].mean())
df
```

# 4)For each of the numeric attributes (proteins upto vitamins), identify and replace all 
noisy data with the median of attribute. 

```python
for i in range(4,12):
    q75 = df.iloc[:,i].quantile(0.75)
    q25 = df.iloc[:,i].quantile(0.25)
    IQR = q75 - q25
    m =   df.iloc[:,i].median()
    high_range = (IQR * 1.5) + q75
    if(q25 > (IQR * 1.5)):
        low_range =  q25 - (IQR * 1.5)
    elif(q25 < (IQR * 1.5)):
        low_range =  (IQR * 1.5) - q25 # outliers lie above high_range and below low_range
        
    for j in range(0,75):
        if(df.iloc[j,i] > high_range) or (df.iloc[j,i] < low_range):
            df.iloc[j,i] = m
        else:
            continue
df
```
