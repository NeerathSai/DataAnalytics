# 5)Which month of the year were least number movie releases are seen? (Note: Extract

# 7)Which are the top 10 movies with maximum return on investment (ROI)? Calculate 
return on investment (ROI) as (BoxOfficeCollection â€“ Budget) / Budget

```python
df['ROI'] = (df['BoxOfficeCollection'] - df['Budget']) / (df['Budget'])
df
```

```python
df['ROI'].max()
```

```python
df.nlargest(10,columns='ROI')
```

# 8)Do the movies have higher ROI if they get released on festive seasons or holiday 
season? Calculate the average ROI for different release times

```python
#YES,movies have higher ROI if they get released on festive seasons or holiday.
df.groupby('ReleaseTime')['ROI'].mean()
```
